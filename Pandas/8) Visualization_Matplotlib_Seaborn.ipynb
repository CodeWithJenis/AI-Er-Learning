```python
import pandas as pd
df_sales = pd.read_excel(r"D:\Carrer\AI\AI Er Learning\Pandas\chapter_3_assets\8_matplotlib_seaborn\linechart.xlsx")
df_sales
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'],df_sales['Fridge'],label="Fridge")
plt.plot(df_sales['Quarter'],df_sales['Dishwasher'],label="Dishwasher")
plt.plot(df_sales['Quarter'],df_sales['Washing Machine'],label="Washing Machine")
plt.title("Product Sales")
plt.legend()
plt.xlabel("Quarter")
plt.ylabel("Revenue")
plt.show()
```

```python
total_sales=df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
plt.pie(total_sales,labels=total_sales.index,autopct="%.2f%%",explode=(0.1,0,0),shadow=True,startangle=140)
plt.show()
```

```python
df_sales.plot(kind="bar",x="Quarter")
plt.xticks(rotation = 45)
plt.show()
```
