```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"D:\Carrer\AI\AI Er Learning\ML\1) Linear_Regression_Single_Variable\data\homeprices.csv")
df
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df['price'])
```

```python
plt.scatter(df["area"],df["price"],color="red",marker="+")
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.plot(df["area"],model.predict(df[['area']]),color="blue")
```

```python
model.coef_
```

```python
model.intercept_
```

```python
predict_df = pd.DataFrame(
    {
    'area':[5000]
    }
)
model.predict(predict_df[['area']])
```
