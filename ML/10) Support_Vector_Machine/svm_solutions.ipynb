```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
df.head()
```

```python
df['target'] = digits.target
df
```

```python
X = df.drop(['target'],axis='columns')
X.head()
```

```python
y = df['target']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
```

```python
# rbf model
model = SVC(kernel='rbf')
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
# linear model
model = SVC(kernel='linear')
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
# so 1st iteration rbf won
# 2nd iteration rbf won
# 3rd iteration rbf won 

# so rbf kernel is better than linear
```

```python
# rbf model (both low)
model = SVC(kernel='rbf',C=0.1,gamma=0.1)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
# rbf model (low,middle)
model = SVC(kernel='rbf',C=0.1,gamma=1)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python

model = SVC(kernel='rbf',C=0.1,gamma=10000)
model.fit(X_train,y_train)
model.score(X_test,y_test)

# increasin gamma has negative impact as overfitiing 
```

```python
# rbf model (middle,middle)
model = SVC(kernel='rbf',C=1,gamma=0.001)
model.fit(X_train,y_train)
model.score(X_test,y_test)

# gamma is decreased to 0.001 gives 100 percent accuarcy also increase in C decrease accuracy from C = 1 also decreasing C makes very much neagtive to the model
```
