```python
import pandas as pd
df = pd.read_csv(r"D:\Carrer\AI\AI Er Learning\ML\14) Naive Bayes Classifier Algorithm\data\titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Embarked','Ticket','Cabin'],axis='columns',inplace=True)
df.head()
```

```python
target = df['Survived']
inputs = df.drop(['Survived'],axis='columns')
```

```python
dummies = pd.get_dummies(inputs['Sex'])
dummies.head()
```

```python
inputs
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
inputs.drop(['Sex'],axis='columns',inplace=True)
inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(inputs,target,train_size=0.8)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
