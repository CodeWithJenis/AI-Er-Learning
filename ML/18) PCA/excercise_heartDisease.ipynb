```python
'''Download heart disease dataset heart.csv in Exercise folder and do following, 
(credits of dataset: https://www.kaggle.com/fedesoriano/heart-failure-prediction)

1) Load heart disease dataset in pandas dataframe
2) Remove outliers using Z score. Usual guideline is to remove anything that has Z score > 3 formula or Z score < -3
3) Convert text columns to numbers using label encoding and one hot encoding
4) Apply scaling
5) Build a classification model using various methods (SVM, logistic regression, random forest) and check which model gives you the best accuracy
6) Now use PCA to reduce dimensions, retrain your model and see what impact it has on your model in terms of accuracy. 
Keep in mind that many times doing PCA reduces the accuracy but computation is much lighter and 
that's the trade off you need to consider while building models in real life
'''
```

```python
import pandas as pd

df = pd.read_csv(r"D:\Carrer\AI\AI Er Learning\ML\18) PCA\data\heart.csv")
df
```

```python
df.describe()
```

```python
df['zBP'] = (df['RestingBP']-df['RestingBP'].mean())/df['RestingBP'].std()
df
```

```python
df['zCholesterol'] = (df['Cholesterol']-df['Cholesterol'].mean())/df['Cholesterol'].std()
df
```

```python
df['zBS'] = (df['FastingBS']-df['FastingBS'].mean())/df['FastingBS'].std()
df
```

```python
df['zHR'] = (df['MaxHR']-df['MaxHR'].mean())/df['MaxHR'].std()
df
```

```python
df['zOldpeak'] = (df['Oldpeak']-df['Oldpeak'].mean())/df['Oldpeak'].std()
df
```

```python
df2 = df[(df['zBP']>-3) & (df['zBP']<3) & (df['zCholesterol']>-3) & (df['zCholesterol']<3) & (df['zBS']>-3) & (df['zBS']<3) & (df['zHR']>-3) & (df['zHR']<3) & (df['zOldpeak']>-3) & (df['zOldpeak']<3)]
df2
```

```python
df.shape[0]-df2.shape[0]
```

```python
df2 = df2.drop(columns=['zBP','zCholesterol','zBS','zHR','zOldpeak'])
df2 
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dfle = df2.copy()
dfle['Sex'] = le.fit_transform(dfle['Sex'])
dfle
```

```python
dfle['ExerciseAngina'] = le.fit_transform(dfle['ExerciseAngina'])
dfle
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    transformers=[('ohe',OneHotEncoder(drop='first'),[2,6,10])],
    remainder='passthrough'
)
dfle = ct.fit_transform(dfle)
new_columns = ct.get_feature_names_out()
df_final = pd.DataFrame(dfle,columns=new_columns)
df_final.columns = [col.split("__")[-1] for col in df_final.columns]


```

```python
df_final
```

```python
df_final.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# List of columns to scale
cols_to_scale = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']

# Apply scaling only to these columns
df_final[cols_to_scale] = scaler.fit_transform(df_final[cols_to_scale])

df_final.head()
```

```python
X = df_final.drop(columns=['HeartDisease'])
y = df_final['HeartDisease']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 30)
```

```python
X_train
```

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train,y_train)
model_svc.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression

model_logistic = LogisticRegression()
model_logistic.fit(X_train,y_train)
model_logistic.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_randomForest = RandomForestClassifier()
model_randomForest.fit(X_train,y_train)
model_randomForest.score(X_test,y_test)
```

```python
# Logistic is the best
from sklearn.decomposition import PCA

pca = PCA(0.95)
x_pca = pca.fit_transform(X)

```

```python
pca.n_components
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
x_pca
```

```python
X_pca_train,X_pca_test,y_train,y_test = train_test_split(x_pca,y,test_size=0.2,random_state=2)
X_pca_train
```

```python
model_logistic.fit(X_pca_train,y_train)
model_logistic.score(X_pca_test,y_test)
```
