```python
import pandas as pd
df = pd.read_csv(r"D:\Carrer\AI\AI Er Learning\ML\13) K_Means_Clustering\data\income.csv")
df
```

```python
from matplotlib import pyplot as plt

plt.scatter(df['Age'],df['Income($)'])
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 3)
km
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df['cluster']==0]
df2 = df[df['cluster']==1]
df3 = df[df['cluster']==2]

plt.scatter(df1['Age'],df1['Income($)'],color='green',label='cluster1')
plt.scatter(df2['Age'],df2['Income($)'],color='red',label='cluster2')
plt.scatter(df3['Age'],df3['Income($)'],color='black',label='cluster3')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler =MinMaxScaler()
```

```python
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)

y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df['cluster']==0]
df2 = df[df['cluster']==1]
df3 = df[df['cluster']==2]

plt.scatter(df1['Age'],df1['Income($)'],color='green',label='cluster1')
plt.scatter(df2['Age'],df2['Income($)'],color='red',label='cluster2')
plt.scatter(df3['Age'],df3['Income($)'],color='black',label='cluster3')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='blue',marker='+',label='centroid')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
km.cluster_centers_
```

```python
sse = []
for k in range(1,10):
    km = KMeans(n_clusters = k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
sse
```

```python
plt.plot(range(1,10),sse)
```
