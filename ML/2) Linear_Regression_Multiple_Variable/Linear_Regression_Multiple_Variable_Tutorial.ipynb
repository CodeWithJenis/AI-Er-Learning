```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv(r"D:\Carrer\AI\AI Er Learning\ML\2) Linear_Regression_Multiple_Variable\data\homeprices.csv")
df
```

```python
median_bedrooms = df['bedrooms'].median()
median_bedrooms
```

```python
df['bedrooms'] = df['bedrooms'].fillna(median_bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df['price'])
print(reg.coef_)
print(reg.intercept_)
```

```python
predict_df = pd.DataFrame({
    'area': [3000,2500],
    'bedrooms': [3,4],
    'age': [40,5]
}
)
predict_df
```

```python
reg.predict(predict_df[['area','bedrooms','age']])
```
