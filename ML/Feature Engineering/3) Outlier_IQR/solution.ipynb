```python
'''
Exercise: Outlier Detection and Removal Using IQR
You are given height_weight.csv file which contains heights and weights of 1000 people. 
Dataset is taken from here, https://www.kaggle.com/mustafaali96/weight-height

You need to do this,

(1) Load this csv in pandas dataframe and first plot histograms for height and weight parameters

(2) Using IQR detect weight outliers and print them

(3) Using IQR, detect height outliers and print them
'''
```

```python
import pandas as pd

df = pd.read_csv(r"D:\Carrer\AI\AI Er Learning\ML\Feature Engineering\3) Outlier_IQR\data\height_weight.csv")
df
```

```python
# (1) Load this csv in pandas dataframe and first plot histograms for height and weight parameters

from matplotlib import pyplot as plt

plt.hist(df['height'],bins=15,rwidth=0.8)
plt.xlabel('Height')
plt.ylabel('Count')
```

```python
df.describe()
```

```python
plt.hist(df['weight'],bins=15,rwidth=0.8)
plt.xlabel('Weight')
plt.ylabel('Count')
```

```python
Q1_Weight = df['weight'].quantile(0.25)
Q3_Weight = df['weight'].quantile(0.75)
Q1_Weight,Q3_Weight
```

```python
IQR_Weight = Q3_Weight-Q1_Weight

lower_limit_Weight = Q1_Weight-1.5*IQR_Weight
upper_limit_Weight = Q3_Weight+1.5*IQR_Weight

lower_limit_Weight,upper_limit_Weight
```

```python
# (2) Using IQR detect weight outliers and print them

df[(df['weight']<lower_limit_Weight) | (df['weight']>upper_limit_Weight)]
```

```python
Q1_Height = df['height'].quantile(0.25)
Q3_Height = df['height'].quantile(0.75)
Q1_Height,Q3_Height
```

```python
IQR_Height = Q3_Height-Q1_Height

lower_limit_Height = Q1_Height-1.5*IQR_Height
upper_limit_Height = Q3_Height+1.5*IQR_Height

lower_limit_Height,upper_limit_Height
```

```python
# (3) Using IQR, detect height outliers and print them

df2 = df[(df['height']<lower_limit_Height) | (df['height']>upper_limit_Height)]
```
